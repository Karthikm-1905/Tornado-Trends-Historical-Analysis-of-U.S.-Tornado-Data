# Q1. How many tornadoes were there per year?

Explanation:- To calculate the number of tornadoes each year, group the years first and then find the length of the column "om" to get the total number of tornadoes. This gives us the annual number of tornadoes. I'm using the group by function to group the data by year, as well as the agg() method to get the length.

```python
tornadoesPerYear = data.groupby("yr").agg({'om':len})
```

```python
tornadoesPerYear
```

```python
tornadoesPerYear.plot(kind='bar',figsize=(24,8))
```

## Q2) What month had the most tornadoes? (Bar Chart)

Explanation:- To determine the month with the most tornadoes, arrange the data by month and then calculate the length of "om," which represents the number of tornadoes in a given state. We may deduce from the bar graph and table below that the month 5 (May) had the most tornadoes.

```python
maxTornados = data.groupby("mo").agg({'om':len})
```

```python
maxTornados.plot(kind='bar',figsize = (22,8))
```

```python
maxTornados
```

May month has the most Tornadoes.

## Q3) What state had the most tornados? (bar chart)

## Q4) What states had the most F4 tornados? (count)

Explanation:- To determine which states experienced the most F4 tornadoes, I first constructed a new data frame by deleting any columns with the value "f4" 0. First, we must sort the data by state and determine the length of "om." This tells us the total number of tornadoes that have occurred in each state. Then, in decreasing order, sort it. We may retrieve the top state with the highest f4 count by utilizing the head() method. We may deduce from the graph and table below that the state "tx" has the highest number of tornadoes.

```python
dt = data[data["f4"] > 0]
```

```python
maxF4Tornadoes =dt.groupby("st").agg({'f4':len}).sort_values("f4", ascending = False)
```

```python
maxF4Tornadoes
```

## Q6) Does there appear to be a typical time of day for tornados to strike? (plot)

Explanation:- I used a line chart to figure out when tornadoes were the most common. Data must first be sorted by time, and then the length of the column "om" must be determined. This generates a table containing the time and number of torandos that occurred at that moment. We may deduce from the graph and table that the highest number of tornadoes occurred between 15:00 and 17:30.
